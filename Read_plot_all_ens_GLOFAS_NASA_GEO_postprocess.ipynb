```python

# Set station of interest manually
index_ = 0 # 3, 7, 8, 9 

name = 'GEO'

# hec = 0, 1, 6, 10, 
# sem obs = 2, 4, 5


if index_ == 0: 
    x_sup = 30000
elif index_ == 3: 
    x_sup = 2500
elif index_ == 6: 
    x_sup = 18000
elif index_ == 7: 
    x_sup = 4000
elif index_ == 8: 
    x_sup = 1500
elif index_ == 9: 
    x_sup = 500
elif index_ == 10: 
    x_sup = 10000

station_name = info_stations['Name'][index_]
station_code = info_stations['Code'][index_]

print('----------------------------------------------------------------------------')

print(station_code)
print(station_name)
    

df_obs = read_station('Telemetry', station_code)

# Convert 'date' to datetime if it's not already
df_obs['date'] = pd.to_datetime(df_obs['date'])

# Set 'date' as the index of the DataFrame
df_obs.set_index('date', inplace=True)

# Display the first few rows of the DataFrame
print(df_obs.head())
```

```python


if name == 'Glofas':

    with open(f'01_GloFAS/data_glofas_{station_code}.pkl', 'rb') as file:
        processed_data = pickle.load(file)
    members = 51

elif name == 'NASA':
        
    with open(f'04_NASA/GHI-MR-MOGREPS-G-hymap-ROUTING/data_nasa_{station_code}.pkl', 'rb') as file:
        processed_data = pickle.load(file)
    members = 18

elif name == 'GEO':
        
    with open(f'05_GEOGLOWS/data_geo_{station_code}.pkl', 'rb') as file:
        processed_data = pickle.load(file)
    members = 18
```

```python

# Assume 'read_dataset' function is already defined and available here

n = 0
# Loop through all dates in the GloFAS data
for issued_date, data_nasa in processed_data.items():
    
    n = n+1

    issued_date = issued_date.normalize()  # Normalize to midnight for date-only comparison
    print(issued_date)

    # Use the read_dataset function to handle each type of data
    dis24_station_nasa = read_ensemble(processed_data, issued_date, 'NASA')


    # Assuming dis24_station_nasa is a pandas DataFrame
    dis24_station_nasa["p5"] = np.percentile(dis24_station_nasa, 5, axis=1)
    dis24_station_nasa["p25"] = np.percentile(dis24_station_nasa, 25, axis=1)
    dis24_station_nasa["median"] = np.percentile(dis24_station_nasa, 50, axis=1)
    dis24_station_nasa["p75"] = np.percentile(dis24_station_nasa, 75, axis=1)
    dis24_station_nasa["p95"] = np.percentile(dis24_station_nasa, 95, axis=1)


    plt.figure(figsize=(3, 4))

    # plt.plot(dis24_station_nasa, label='ensemble nasa', color='grey', alpha = 0.5)  # Use solid line for other datasets
    plt.fill_between(dis24_station_nasa.index, dis24_station_nasa['p5'], dis24_station_nasa['p95'], color='grey', alpha=0.3, linewidth=0, label='5th-95th percentile band')
    plt.fill_between(dis24_station_nasa.index, dis24_station_nasa['p25'], dis24_station_nasa['p75'], color='grey', alpha=0.5, linewidth=0, label='25th-75th percentile band')
    plt.plot(df_obs[name_data], color='black',linestyle='--',linewidth=1, label='Obs')
    plt.plot(dis24_station_nasa['median'], label='Median', color='blue', linewidth=1)

    
    plt.xlabel('Date')
    plt.ylabel('Discharge (m³/s)')
    plt.title(f'Issued on {issued_date.date()}')

    plt.xlim(issued_date - timedelta(days=1), issued_date + timedelta(days=7))

    # plt.ylim(min(df_obs[name_data]), x_sup)
    # plt.ylim(min(dis24_mean_glofas['dis24_station']), 1.1*max(dis24_mean_glofas['dis24_station']))

    plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=1))
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m/%d'))
    plt.gcf().autofmt_xdate(rotation=90, ha='center')

    plt.grid(True, color='grey', alpha=0.2, which='both', linestyle='-', axis='x')
    plt.tight_layout()
    plt.axvline(x=issued_date, color='grey', label='Issued Date')
    # plt.legend(loc='lower left', frameon=True, facecolor='white', edgecolor='white')

    plt.savefig(f'Figures/{station_name}_forecast_{name}_ENS_{n}.png', dpi=300)
    plt.show()
```
